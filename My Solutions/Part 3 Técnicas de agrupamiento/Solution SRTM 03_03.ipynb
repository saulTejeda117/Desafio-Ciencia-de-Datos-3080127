```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import normalize
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('clientes.csv', delimiter=";")
print(df.head(10))
```

```python
# Creamos un dendrograma utilizando el método 'ward' para el enlace y la métrica 'euclidean' para la distancia.
dendrogram(linkage(df, method='ward', metric='euclidean'))

# Agregamos una línea horizontal en el dendrograma en el punto y=15000, con color rojo y estilo de línea discontinua.
plt.axhline(y=15000, color='r', linestyle='--')
```

```python
# Creamos una instancia del algoritmo AgglomerativeClustering con 3 clusters
agglomerative = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

# Aplicamos el algoritmo de agrupamiento a los datos en el DataFrame 'df'
agglomerative.fit(df)

# Obtenemos las etiquetas de los clusters asignados a cada punto de datos
labels = agglomerative.labels_
```

```python
print(labels)
```

```python
# Crear un gráfico de dispersión para visualizar la relación entre el salario y el gasto
# Utilizamos los datos del DataFrame df y asignamos colores según la variable 'labels'
plt.scatter(df['Salario'], df['Gasto'], c=labels) 

# Etiquetas para los ejes del gráfico
plt.xlabel('Salario Mensual')
plt.ylabel('Porcentaje Gasto Mensual')
```
