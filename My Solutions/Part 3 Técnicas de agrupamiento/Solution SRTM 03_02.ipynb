Imprime las coordenadas del centroide de cada grupo. Grafica los resultados y los centros de las agrupaciones determinados por el estimador de k-means."

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('clusters.csv', delimiter=";")
print(df.head(10))
```

```python
# Creamos una instancia de KMeans con 4 clústeres
kmeans = KMeans(n_clusters=4)

# Ajustamos el modelo de KMeans a los datos en el DataFrame df
kmeans.fit(df)

# Predecimos a qué clúster pertenece cada punto en df
y_kmeans = kmeans.predict(df)

# Obtenemos las coordenadas de los centros de los clústeres
cluster_centers = kmeans.cluster_centers_
```

```python
# Crear una gráfica de dispersión para mostrar los puntos de datos coloreados según sus etiquetas de clúster
plt.scatter(df['Latitud'], df['Longitud'], c=y_kmeans, s=20)  # Scatter plot de puntos de datos coloreados según etiquetas de clúster
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200)  # Scatter plot de centros de clúster en rojo
plt.xlabel('Latitud')  # Etiqueta del eje x
plt.ylabel('Longitud')  # Etiqueta del eje y
```
