"Mostrar los grupos que se forman a partir de los datos que se nos han dado"

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('datos.csv', delimiter=";")
print(df.head(10))
```

```python
# Creamos una instancia de la clase DBSCAN
dbscan = DBSCAN()

# Ajustamos el modelo DBSCAN a los datos en el dataframe df
dbscan.fit(df)

# Obtenemos las etiquetas asignadas a cada punto por el algoritmo DBSCAN
labels = dbscan.labels_
```

```python
print(labels)
```

```python
plt.scatter(df['Variable1'],df['Variable2'], c=labels, cmap= "plasma")
```
